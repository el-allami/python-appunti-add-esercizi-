```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify as sqf  ## per importare
```

a. crei un dataframe contenente le informazioni riportate nella tabella sottostante, relativa al numero di ordini (order_numbers) seguiti da vari commessi (last_name) di un negozio.

```python


d = {
    "last_name": ["Boyer","CopeLand","Daniel","Serrano","Terrell","Vargas"],
    "order_numbers": [185,50,196,58,24,28]
    }

df = pd.DataFrame(d)
df
```

b. visualizzi un grafico a barre verticali per rappresentare i dat. Il risultato deve essere questo:

```python

fig = plt.figure(figsize = (8,8))
ax = plt.axes()

ax.bar(df["last_name"], df["order_numbers"])
```

c. visualizzi lo stesso grafico ma con altri colori delle barre, un titolo appropriato al grafico e i nomi degli
assi.

```python
fig = plt.figure(figsize = (8,8))
ax = plt.axes()

plt.rcParams.update({'font.size': 10})

ax.set_xlabel('last names', fontsize=20)
ax.set_ylabel('order numbers', fontsize=20)
ax.bar(df["last_name"], df["order_numbers"], color = "g")

fig.suptitle('Statistics of restaurants', color='k', fontsize=30)
plt.show()
```

d. Visualizzare lo stesso grafico ma con le barre in ordine decrescente

```python
################          COME CREARE UN GRAFICO IN ORDINE DECRESCENTE



x = list(df.sort_values("order_numbers",ascending=False)["last_name"])
y = list(df["order_numbers"].sort_values(ascending=False))

fig = plt.figure(figsize = (8,8))
ax = plt.axes()

plt.rcParams.update({'font.size': 10})

ax.set_xlabel('last names', fontsize=20)
ax.set_ylabel('order numbers', fontsize=20)
ax.bar(x, y, color = "g")

fig.suptitle('Statistics of restaurants', color='k', fontsize=30)
plt.show()
```

e. visualizzi un grafico a barre orizzontali per rappresentare i dati.

```python
################          COME CREARE UN GRAFICO (ISTOGRAMMA) IN MODO ORRIZZONTALE



x = list(df["last_name"])
y = list(df["order_numbers"])

fig = plt.figure(figsize = (8,8))
ax = plt.axes()

plt.rcParams.update({'font.size': 10})

ax.set_ylabel('last names', fontsize=20)
ax.set_xlabel('order numbers', fontsize=20)
ax.barh(x, y, color = "g")  ### si usa barh   ---->  h sta per horizontal

fig.suptitle('Statistics of restaurants', color='k', fontsize=30)
plt.show()
```

f. crei un nuovo dataframe contenente la percentuale di ordini consegnati seguiti da ogni componente
dello staff. Visualizzare il cognome del componente dello staff e la percentuale di ordini. Ordinare in
ordine alfabetico sul cognome del componente dello staff.

```python
tot = 185+50+196+58+24+28

d2 = {
    "last_name": ["CopeLand","Serrano","Boyer","Daniel","Vargas","Terrell"],
    "order_perc": [(50*100)/tot,(58*100)/tot,(185*100)/tot,(196*100)/tot,(28*100)/tot,(24*100)/tot]
    }
    
df2 = pd.DataFrame(d2)
df2
```

g. visualizzi un grafico a torta per rappresentare i dati estratti.

```python
perc = list(df2["order_perc"])
nomi = list(df2["last_name"])

fig = plt.figure(figsize = (10,10))
ax = plt.axes()
plt.rcParams.update({'font.size' : 15})
ax.pie(perc,labels = nomi, autopct="%1.2f%%")
plt.show
```

h. visualizzi lo stesso grafico ma con altri colori delle percentuali e un titolo appropriato al grafico

```python
perc = list(df2["order_perc"])
nomi = list(df2["last_name"])
colors = ["m","c","r","g","b","w"]

fig = plt.figure(figsize = (10,10))
ax = plt.axes()
plt.rcParams.update({'font.size' : 20})
ax.pie(perc,labels = nomi, colors = colors, autopct="%1.2f%%")  #  colors = -----> si usa per cambiare il colore delle fette
fig.suptitle('Grafico a torta', color='k', fontsize=30)
plt.show
```

i. rappresentare le stesse informazioni utilizzando come tipo di grafico il TreeMap (google it)

```python
size = list(df2["order_perc"])
nomi = list(df2["last_name"])
sqf.plot(sizes = size, label = nomi, alpha = 0.6)
plt.axis("off")
plt.show
```
