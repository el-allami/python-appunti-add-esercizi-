a. Scrivere le istruzioni per caricare i dati del file csv (inventare il link e il nome del file) in un dataframe Pandas chiamato “movies”


```python
import pandas as pd

url = "https://gist.githubusercontent.com/tiangechen/b68782efa49a16edaf07dc2cdaa855ea/raw/0c794a9717f18b094eabab2cd6a6b9a226903577/movies.csv"

movies = pd.read_csv(url)
movies
```

b. visualizzare le ultime 8 righe del dataframe


```python
movies.tail(8) 
```

c. visualizzare solo le colonne che contengono il titolo e il genere del film


```python
movies.loc[:,['Film', 'Genre']]
```

d. visualizzare le righe del dataframe dalla 20 alla 30


```python
movies.iloc[20:31]
```

e. visualizzare i titolo e l’anno dei film che si trovano nelle posizioni 23, 29 e 45


```python
movies.loc[[23,29,45],["Film", "Year"]]           ########    PER TROVARE COLONNE SPECIFICHE DI ELEMENTI SPECIFICI
```

f. visualizzare i titoli dei film del genere “Animation”


```python
movies.loc[movies["Genre"] == "Animation", ["Film", "Genre"]]
```

g. visualizzare i titoli dei film prodotti dalla “Disney” e dalla “20th Century Fox”


```python
movies.loc[(movies["Lead Studio"] == "Disney") | (movies["Lead Studio"] == "20th Century Fox"), ["Film", "Lead Studio"]]
```

h. visualizzare i titoli dei film usciti dal 2005 al 2010


```python
movies.loc[(movies["Year"] <= 2010) & (movies["Year"] >= 2005), ["Film", "Year"]]
```

i. visualizzare i film (tutte le colonne) il cui titolo comincia con la lettera W


```python
movies[movies["Film"].str.startswith("W")]          #######      UN MODO PER VISALIZZARE LE COLONNE CON CUI LA PRIMA COLONNA INIZIA CON UNA LETTERA SPECIFICA
```

```python

movies[movies["Film"].str.contains("for")]    ###### per vedere se contiene per esempio for
```

```python

movies[movies["Film"].str.endswith("s")]    ###### per vedere se finisce con s
```

j. visualizzare l’ “Audience score %” medio dei film del genere “Drama”


```python
movies.loc[movies["Genre"] == "Drama",["Audience score %"]].mean()
```

k. visualizzare il numero di film per ogni genere

```python
movies.groupby("Genre")["Film"].count()
```

l. diminuire del 10% il valore delle recensioni presenti nel campo “Audience score %”


```python
movies["Audience score %"] = movies["Audience score %"].apply(lambda x: x * 0.9)
movies
```

```python
movies["Audience score %"] = movies["Audience score %"] * 0.9
movies
```
