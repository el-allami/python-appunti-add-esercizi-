# Index del dataframe
E' possibile impostare come valore dell'index del *DataFrame* i valori di una colonna specifica o un Array avente un numero di valori pari a quello delle righe

Ci sono due modi per impostare l'index del *DataFrame*:
* Passandogli come valore il nome di una colonna esistente `DataFrame.set_index('column_name' or list)` | [Docs](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.set_index.html)
* Impostandalo direttamente alla creazione di esso

Come dimostreremo in questo Notebook è consigliabile impostare un index alle righe per identificarle al meglio e per ridurre il tempo di esecuzione qunado ne viene avviata la ricerca.

```python
import pandas as pd

df = pd.read_csv('./data/pokemon_list.csv')

df
```

```python
%%time
for _ in range(1, 10000): 
  df.loc[df['Name'] == 'Pikachu']


df.loc[df['Name'] == 'Pikachu']
```

```python
df = df.set_index('Name')

df
```

```python
%%time

for _ in range(1, 10000): 
  df.loc['Pikachu']


df.loc['Pikachu']
```
